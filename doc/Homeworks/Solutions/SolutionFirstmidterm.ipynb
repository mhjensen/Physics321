### Part 1, Particle in a one-dimensional  potential

We consider a particle (for example an atom) of mass $m$ moving in a one-dimensional potential,

$$
V(x)=\frac{V_0}{d^4}\left(x^4-2x^2d^2+d^4\right).
$$

We will assume all other forces on the particle are small in comparison, and neglect them in our model.  The parameters $V_0$ and $d$ are known constants. 

1. (5pt) Plot the potential and find the  equilibrium points (stable and unstable) by requiring that the first derivative of the potential is zero. Make an energy diagram (see for example Malthe-Sørenssen chapter 11.3) and mark the equilibrium points on the diagram and characterize their stability. The position of the particle is $x$. 

```python
# arbitrary constants
V0 = 1
d = 20

x = np.arange(-40,40.1,0.1)
V = V0*(x**4 - 2*(x**2)*d**2 + d**4)/d**4
stable = np.array([np.array([-20,20]),np.array([0,0])])
unstable = np.array([np.array([0]), np.array([1])])

plt.figure(figsize=(10,5))
plt.scatter(stable[0], stable[1], label='Stable equilibrium points', c='orange', s=50)
plt.scatter(unstable[0], unstable[1], label='Unstable equilibrium points', c='g', s=50)
plt.plot(x, V)
plt.title('1D Potential')
plt.xlabel('X-position (m)')
plt.ylabel('Potential (J)')
plt.grid()
plt.legend()
plt.show()
```

5. (5pt) Use the above potential to set up the total forces acting on the particle.  Find the acceleration acting on the particle. Is this a conservative force? Calculate also the **curl** of the force  $\boldsymbol{\nabla}\times \boldsymbol{F}$ in order to validate your conclusion. 

* Total forces on the particle: $F^{net} = -\boldsymbol{\nabla}V$
* An expression for the force can be found as:

$$
F^{net} = -\boldsymbol{\nabla}V(x) = -\left( \frac{d}{dx}V(x) \right)
$$

$$
= -\frac{d}{dx} \left( \frac{V_0}{d^4}\left(x^4 - 2x^2d^2 + d^4\right) \right)
$$

$$
F(x) = - \frac{4V_0}{d^4}\left(x^3 - xd^2\right) 
$$

* Acceleration can be found using the relationship $F_x = ma_x$:

$$
F(x) = ma(x)
$$

$$
a(x) = \frac{F(x)}{m} = -\frac{1}{m}\frac{4V_0}{d^4}\left(x^3 - xd^2\right) 
$$

* The potential function is not time-dependent, i.e. only depends on position, which indicates that the associated force is conservative. 
* Since this particle obeys a 1-dimensional (x) potential (and therefore 1-dimensional force) function, the curl of the force is found as follows:

$$
\boldsymbol{\nabla} \times F = (0 - 0)\boldsymbol{i} - (0 - 0)\boldsymbol{j} + (0 - 0)\boldsymbol{k}
$$

$$= 0$$
* Since the curl of the force is zero, this validates that the force is conservative.

6. (5pt) Are linear momentum and angular momentum conserved? You need to show this by calculating the quantities.

* For linear momentum to be conserved, there must be zero change in momentum relative to time $t$. The change in momentum $p$ can be defined as:

$$
\frac{dp}{dt} = F^{ext} = - \frac{4V_0}{d^4}\left(x^3 - xd^2\right) 
$$

* where $F^{ext}$ is the net force on the particle found in #5. This force varies with position, so when the force is nonzero, i.e. $\frac{dp}{dt}$ equals a nonzero amount, linear momentum is not conserved. Linear momentum is only conserved at positions where the net external force equals zero, which are calculated as $x = 0$ and $x = \pm \sqrt{d}$.

* For angular momentum to be conserved, the change in angular momentum (or net torque) on the particle must be zero. Angular momentum $\boldsymbol{L}$ is defined as:

$$
\boldsymbol{L} \;=\; \boldsymbol{p} \times \boldsymbol{r}
$$

* Because $\frac{dp}{dt} = F^{ext}$, the change in $\boldsymbol{L}$, or torque, is defined as:

$$
\tau \;=\; \frac{d\boldsymbol{L}}{dt} \;=\; \boldsymbol{F} \times \boldsymbol{r}
$$

$$
=\; (F(x),\; 0,\; 0) \times (x,\; 0,\; 0) \;=\; (0,\; 0,\; 0)
$$

* Because there is no torque, there is no change in angular momentum, and $\boldsymbol{L}$ is conserved.

7. (10pt) Write a numerical algorithm to find the position and velocity of the particle at a time $t+\Delta t$ given the position and velocity at a time $t$. Here you can use either the standard forward Euler, or the Euler-Cromer or the Velocity Verlet algorithms.   You need to justify your choice here (hint: consider energy conservation).

* To calculate the position and velocity from this acceleration, the Velocity Verlet algorithm will be used to ensure that the conservation of energy associated with this force is represented.

```python
# time bounds in s
t0 = 0 
tf = 30
dt = 0.0001
t = np.arange(t0,tf+dt,dt)

m = 1 # mass in kg
V0 = 1 # potential E in J
d = 0.1 # constant in m

x0 = d # initial position in m
# initial velocities in m/s
v0a = 0.5 
v0b = 1.5 
v0c = 2.5

def potential(t, r0, v0, a_eq):
    r = np.zeros(len(t))
    v = np.zeros(len(t))

    r[0] = r0
    v[0] = v0

    # calculate acceleration
    for i in range(len(t)-1):
        a_i = a_eq(r[i])
        r[i+1] = r[i] + dt*v[i] + 0.5*a_i*(dt**2)
        
        a_ip1 = a_eq(r[i+1])
        v[i+1] = v[i] + 0.5*dt*(a_i + a_ip1)
    
    return r, v

def a1D(x):
    accel = -4*V0*x*(x**2 - d**2) / (m*d**4)
    return accel
```

### Part 2, a potential with confinement

In the final part we are going to study a new potential. This potential contains the basic mathematical components needed to describe the confinement of quarks through a term $\kappa x$. The aim of this exercise is to try to develop your intuition about the motion of physical objects due to specific forces. Then we will test our intution by running simulations. The potential, in one dimension only is defined for $x\in [0.2,\infty)$.  It reads

$$
V(x) = -\frac{\gamma}{x}+\frac{\delta}{x^2}+\kappa x,
$$

where the last term is the one which ensures confinement of for example quarks.

1. (5pt) Plot the potential for $x\in [0.2,10]$ and set $\gamma=10$, $\delta = 3$ and $\kappa =1$. Find the value of $x$ where the potential has a minimum.

```python
gm = 10
dl = 3
kp = 1

x = np.arange(0.2, 10.001, 0.001)
V = -(gm/x) + (dl/(x**2)) + kp*x

plt.plot(x, V)
plt.xlabel('x-position (m)')
plt.ylabel('Potential (J)')
plt.title('1D Quark Confinement Potential')
plt.grid()
plt.show()
```

* Over this interval, the potential has a minimum at x-values where the first derivative of potential is 0.
* Setting the derivative equal to zero and plugging in the constants' given values:

$$
\frac{d}{dx} \left(-\frac{\gamma}{x}+\frac{\delta}{x^2}+\kappa x \right) \;=\;
\frac{\gamma}{x^2} - \frac{2\delta}{x^3} + \kappa \;=\; 0
$$

$$
\frac{10}{x^2}\left(\frac{x}{x}\right) - \frac{6}{x^3} + 1\left(\frac{x^3}{x^3}\right)  = 0
$$

$$
\frac{x^3 + 10x - 6}{x^3} = 0 \; \rightarrow \;
x^3 + 10x - 6 = 0
$$

* Using Wolfram Alpha to solve for the position of minimum potential energy on this interval:

$$
x^3 + 10x - 6 = 0 \;\;\rightarrow \;\; x = 0.58044 \;m
$$

2. (5pt) Show that this potential leads to an energy conserving force by calculating the **curl** of the resulting force.

* The force can be found as:

$$
F^{net} = -\boldsymbol{\nabla}V(x) = -\left( \frac{d}{dx}V(x), \frac{d}{dy}V(x), \frac{d}{dz}V(x) \right) \;=\; -\frac{d}{dx} \left( -\frac{\gamma}{x}+\frac{\delta}{x^2}+\kappa x, \right)
$$

$$
F^{net} \;=\; -\frac{\gamma}{x^2} + \frac{2\delta}{x^3} - \kappa
$$

* The curl of the force is then:

$$
\boldsymbol{\nabla} \times \boldsymbol{F} = \left( \frac{d}{dx},\; \frac{d}{dy},\; \frac{d}{dz} \right) \times \left(\left( -\frac{\gamma}{x^2} + \frac{2\delta}{x^3} - \kappa\right),\; 0,\; 0 \right)
\;=\; (0,\; 0,\; 0)
$$

* Since the resulting curl of the force is 0, the force is conservative.

3. (10pt) Assume now that at $x=2$ the particle moving in this potential is at rest, that is its velocity is zero. Find the total energy and describe what kind of motion you can expect. The point $x=2$ is a so-called turning point. Can you find the other turning point where the kinetic energy is zero. If you keep increasing $x$, will the particle ever be able to escape the potential well?

```python
xturn = 2 # initial position in m

K0 = 0
V0 = -(gm/xturn) + (dl/(xturn**2)) + kp*xturn
E_tot = K0 + V0
print('Total energy =',E_tot,'J')
```

* Observing the graph of potential energy as a function of position, a total energy of -2.25 J will confine the particle to the range of positions that occupy the potential well in between 0 and 2 meters. This indicates that we can expect the motion of the particle to be an oscillation between the two turning points of this potential well, one of which is defined above to be $x=2$. The other turning point will be another location of maximum potential energy (equal to the total energy of the system) and 0 kinetic energy: 

$$
V(x) = E^{max}
$$

$$
V(x) = -\frac{10}{x}+\frac{3}{x^2}+1 x = -2.25
$$

$$
x^2\left( -\frac{10}{x}+\frac{3}{x^2} \right) = x^2\left( -2.25 - x \right)
$$

$$
-10x + 3 = -2.25x^2 - x^3 \;\;\rightarrow \;\; x^3 + 2.25x^2 - 10x + 3 = 0
$$

* Wolfram Alpha solutions for this cubic equation in the given range:

$$
x = 2 \;\;\text{and} \;\; x = 0.328
$$

* If the particle never has an initial velocity greater than zero, then the particle will not be able to escape the potential well, no matter how large the initial position is. This potential leads to a force which describes an acceleration in the negative $x$-direction for positions greater than that of the minimum potential energy, since $F= -\boldsymbol{\nabla}V(x)$ and the potential is increasing at all points on the aforementioned range. Therefore, a particle under this force at positions greater than $x=2$ will always accelerate towards the position of minimum potential and exhibit oscillatory motion.

4. (20pt) Finally we will now change our program (using either the Euler-Cromer or the Velocity-Verlet method) and compute the position and the velocity as functions of time using the force computed from the above potential. Use as initial condition at time $t=0$ that the particle is $x_0=2$ and has initial velocity $v_0=0$. Test numerically that energy is conserved as function of time. And test that the particle moves according to your analysis from the previous exercise.

```python
r0 = 2
v0 = 0
m = 1

t0 = 0
tf = 20
dt = 0.0001
t = np.arange(t0, tf+dt, dt)

def a_quark(r):
    a = (-1/m)*((gm/(r**2)) + (-2*dl/(r**3)) + kp)
    #print(a)
    return a

r_q, v_q = potential(t, r0, v0, a_quark)
```

* To test if energy is conserved, the total energy at every point in time over the interval can be calculated:

```python
# potential energy over time:
V_q = -(gm/r_q) + (dl/(r_q**2)) + kp*r_q

# kinetic energy over time:
K_q = 0.5*m*v_q**2

# total energy over time:
E_q = V_q + K_q

plt.figure(figsize=(10,5))

plt.plot(t, V_q, label='Potential')
plt.plot(t, K_q, label='Kinetic')
plt.plot(t, E_q, label='Total')
plt.xlabel('Time (s)')
plt.ylabel('Energy (J)')
plt.title('Conservation of Energy')
plt.legend()
plt.grid()
plt.show()
```

* The graph of total energy (kinetic + potential) as a function of time is constant, demonstrating that energy is conserved in this simulation.
* Checking for discrepancy between the theoretical total energy of the system and the numerical solution can test that the particle moves according to the analysis from the previous exercise:

```python
# absolute error of energy
E_q_err = np.abs(E_q - E_tot)
plt.figure(figsize=(10,5))

plt.plot(t, E_q_err)
plt.xlabel('Time (s)')
plt.ylabel('|Numerical - Theoretical| (J)')
plt.title('Absolute Error of Energy')
plt.grid()
plt.show()
```

* Given that the discrepancy is on the order of $10^{-6}$, this error indicates that the particle does in fact move according to the previous analysis.

### Integrating Classwork With Research

This opportunity will allow you to earn up to 5 extra credit points on a Homework per week. These points can push you above 100% or help make up for missed exercises.
In order to earn all points you must:

1. Attend an MSU research talk (recommended research oriented Clubs is  provided below)

2. Summarize the talk using at least 150 words

3. Turn in the summary along with your Homework.

Approved talks:
Talks given by researchers through the following clubs:
* Research and Idea Sharing Enterprise (RAISE)​: Meets Wednesday Nights Society for Physics Students (SPS)​: Meets Monday Nights

* Astronomy Club​: Meets Monday Nights

* Facility For Rare Isotope Beam (FRIB) Seminars: ​Occur multiple times a week
