### Exercise 1, Two-body Problems and Conservative Forces (total 35pt)

The relevant material can be found in Taylor chapters 4 and 8. Homework sets 5, 6-9 and the chapters on [forces](https://mhjensen.github.io/Physics321/doc/LectureNotes/_build/html/chapter4.html) and [two-body problems](https://mhjensen.github.io/Physics321/doc/LectureNotes/_build/html/chapter6.html) from the lecture notes may also be of use. 

This exercise and the next are  follow-ups of homework 6. There we studied the so-called
[Lennard-Jones potential](https://en.wikipedia.org/wiki/Lennard-Jones_potential) which is widely used in molecular dynamics
calculations and in the simulations of [quantum liquids](https://journals.aps.org/prb/abstract/10.1103/PhysRevB.40.11070). This potential is based on parameterizations from
experiments. In [molecular dynamics calculations](https://en.wikipedia.org/wiki/Molecular_dynamics) the assumption is
that atoms move according to the laws of Newton, given the correct
model for interactions. We can say then that quantum-mechanical
degrees of freedom stemming from complicated interactions between electrons
and protons in an atom, are parametrized in terms of an effective
potential.

We will limit ourselves to a two-body problem.

The goal of this exercise is to model a gas of argon atoms (here two
atoms only interacting), where the atoms interact according to the
famous Lennard-Jones potential,

<!-- Equation labels as ordinary links -->
<div id="eq:lj"></div>

$$
\begin{equation}
    V(r) = 4\varepsilon\left((\frac{\sigma}{r})^{12} - (\frac{\sigma}{r})^6\right), \label{eq:lj} \tag{1}
\end{equation}
$$

where $r$ is the distance between two atoms,
$r=\vert\boldsymbol{r}_i-\boldsymbol{r}_j\vert$, that is the norm of the relative
distance vector $\boldsymbol{r}$. The quantities $\sigma$ and $\varepsilon$ are
parameters which determine which chemical compound is modelled. This
potential is a good approximation for noble gases like helium, neon, argon and other.

We start first with a basic study of the potential

**1a (5pt):** Plot the potential as a function of $r$ with $\varepsilon=1$ and $\sigma=1$, for example for $r \in [0.9,3]$.

```python
epsilon = 1
sigma = 1
r = np.arange(0.9, 3.01, .01)

V = 4*epsilon*( (sigma/r)**12 - (sigma/r)**6 )

plt.plot(r, V)
plt.title('Change in potential with position')
plt.xlabel('r [m]')
plt.ylabel('V [J]')
plt.grid();
```

**1b (5pt):** The behaviour of $V(r)$ is vastly different for $r < \sigma$ and $r > \sigma$. Which term in the potential, ([1](#eq:lj)), dominates in each case and what is the effect?

* For the case of $r > \sigma$, the term to the sixth power will dominate, and the total energy will be negative (bounded motion for atoms).
* For the case of $r < \sigma$, the term to the twelfth power will dominate, and the total energy will be positive (unbounded motion).

**1c (5pt):** Find and characterize the equilibrium points of the potential.

* The equilibrium points of this potential exist at the values for position where $\frac{dV(r)}{dr} = 0$:

$$
V(r) = 4\varepsilon\left(\left(\frac{\sigma}{r}\right)^{12} - \left(\frac{\sigma}{r}\right)^6\right)
$$

$$
\frac{dV(r)}{dr} \;=\; 
0 \;=\; 
4\varepsilon \left( -12\left( \frac{\sigma^{12}}{r^{13}} \right) + 6\left( \frac{\sigma^{6}}{r^{7}} \right) \right) \;=\; 
\frac{24\varepsilon\sigma^{6}}{r^7} \left( -\frac{2\sigma^{6}}{r^{6}} + 1 \right)
$$

$$
1 - \frac{2\sigma^{6}}{r^{6}} = 0
$$

$$
r = (2\sigma^{6})^{\frac{1}{6}} = (2)^{\frac{1}{6}}\sigma
$$

* Two atoms which start at rest separated by a distance of 1.5$\sigma$ are characterized by an initial (and total) negative system energy. Their relative distance causes them to sit within the "potential well," and their motion will be oscillatory, alternating between their minimum and maximum separation distance.
* Two atoms which start at rest separated by a distance of 0.95$\sigma$ have a positive initial system energy. Their motion is unbounded, because they have enough total energy to escape the potential well and move infinitely away from each other, all of their energy being transfered to kinetic as potential energy goes to zero.

**1f (5pt):**  Are linear and angular momentum conserved? You need to show this by calculating the relevant quantities. Hint: see homework set 4 and exercise 2.

* Linear momentum $\boldsymbol{P}$ for a system of 2 atoms:

$$
\boldsymbol{P} = 
\sum_{i=1}^2 \boldsymbol{p_i} = 
m_1\boldsymbol{v_1} + m_2\boldsymbol{v_2}
$$

* For linear momentum to be conserved, the time derivative of $\boldsymbol{P}$ must be zero:

$$
\frac{d}{dt}\left( m_1\boldsymbol{v_1} + m_2\boldsymbol{v_2} \right)
\;=\;
m_1\boldsymbol{a_1} + m_2\boldsymbol{a_2}
\;=\;
(\boldsymbol{F}^{int}_1 + \boldsymbol{F}^{ext}_1) + (\boldsymbol{F}^{int}_2 + \boldsymbol{F}^{ext}_2)
$$

* If the system is defined such that it can be considered to be isolated, the internal forces in this system are simply the forces of each atom on each other. These internal forces (being equal and opposite) will cancel:

$$
\frac{d}{dt}\left(\boldsymbol{P}\right)
\;=\;
\boldsymbol{F}^{ext}_1 + \boldsymbol{F}^{ext}_2
$$

* The system being isolated also implies that there are no external forces, proving that $\boldsymbol{P} = 0$ and linear momentum is conserved.

* Total angular momentum $\boldsymbol{L}$ of a system of 2 atoms: 

$$
\boldsymbol{L} = 
\sum_{i=1}^2 \boldsymbol{l_i} = 
\boldsymbol{r_1} \times \boldsymbol{p_1} + \boldsymbol{r_2} \times \boldsymbol{p_2}
$$

* For $\boldsymbol{L}$ to be conserved, the $\frac{d}{dt}(\boldsymbol{L})$, or net torque $\boldsymbol{\tau}$, must be zero:

$$
\boldsymbol{\tau} = 
\boldsymbol{r}_1 \times \boldsymbol{F}_1^{net} + \boldsymbol{r}_2 \times \boldsymbol{F}_2^{net}
$$

$$
= \boldsymbol{r}_1 \times \boldsymbol{F_{12}} + \boldsymbol{r}_1 \times \boldsymbol{F_1^{ext}} + \boldsymbol{r}_2 \times \boldsymbol{F_{21}} + \boldsymbol{r}_2 \times \boldsymbol{F_2^{ext}}
$$

* Because the internal forces described are equal and opposite, 

$$
\boldsymbol{\tau} = 
(\boldsymbol{r}_1 - \boldsymbol{r}_2) \times \boldsymbol{F_{12}} + (\boldsymbol{r}_1 \times\boldsymbol{F}_1^{ext} + \boldsymbol{r}_2 \times\boldsymbol{F}_2^{ext})
$$

* Because the system can be defined such that it is isolated from external forces, and the internal forces of the two atoms' interaction are central forces, the cross product representing net torque $\boldsymbol{\tau}$ is 0, and angular momentum is conserved.

**2b (5pt):** What is the characteristic time scale $\tau$, and what is its value for argon, which has $\sigma=3.405$ Å (1 Å = 1E-10 m), $m = 39.95$ u (with 1 u = 1.66E-27 kg) and $\varepsilon$ = 1.0318E-2 eV (1 eV = 1.602E-19 J)?

* The characteristic time scale indicates the order of magnitude the time scale in seconds  which typical behaviors in systems influenced by this potential are expected to occur at. For example, because distance is scaled by length $\sigma$, typical speeds in the system can be assumed to occur at around the order of length $\sigma$ in time $\tau$.

```python
epsilon = (1.0318E-2)*1.602E-19 # units in J
m = 39.95*1.66E-27 # units in kg
sigma = 3.405E-10 # units of m
tau = np.sqrt((m*sigma**2)/epsilon) # characteristic time scale for argon
print('characteristic time scale =',tau)
```

**2c (10pt):** Write a function which solves ([2](#eq:undim)) for two atoms and find the positions and velocities of the atoms as a function of time. Implement either the Euler-Cromer or the Velocity-Verlet methods to solve the equations of motion (you can reuse your codes from homework sets 5-9 or the second midterm).

```python
dt = 0.01
t0 = 0
tf = 5.0
t = np.arange(t0, tf, dt) # units of t'
```

```python
def a(r1i, r2i):
    
    ri = np.abs(r1i - r2i)
    
    # forces are equal and opposite
    a1i = 24*(2*ri**(-12) - ri**(-6))*(r1i - r2i)/ri # accel. due to force on atom 1 by atom 2
    a2i = 24*(2*ri**(-12) - ri**(-6))*(r2i - r1i)/ri # accel. due to force on atom 2 by atom 1
    
    return a1i, a2i


def LJ(r10, r20, v10, v20):
    
    r1 = np.zeros(len(t)) # units of r'
    r2 = np.zeros(len(t))
    
    v1 = np.zeros(len(t)) # units of r'/t'
    v2 = np.zeros(len(t)) 

    r1[0] = r10
    r2[0] = r20
    
    v1[0] = v10
    v2[0] = v20

    for i in range(len(t)-1):
        
        a1_i, a2_i = a(r1i=r1[i], r2i=r2[i])
        
        r1[i+1] = r1[i] + dt*v1[i] + 0.5*(dt**2)*a1_i
        r2[i+1] = r2[i] + dt*v2[i] + 0.5*(dt**2)*a2_i
        
        a1_ip1, a2_ip1 = a(r1i=r1[i+1], r2i=r2[i+1])
        
        v1[i+1] = v1[i] + 0.5*dt*(a1_i + a1_ip1)
        v2[i+1] = v2[i] + 0.5*dt*(a2_i + a2_ip1)
    
    return r1, r2, v1, v2
```

```python
def a(ri_atom, ri_1, ri_2):
    
    rvec_1 = ri_atom - ri_1 # relative position between body and atom 2
    a_1 = 0
    if round(np.linalg.norm(rvec_1),10) != 0:
        norm1 = np.linalg.norm(rvec_1)
        # a due to F from atom 1
        a_1 = 24*( 2*(norm1)**(-12) - (norm1)**(-6) )*(rvec_1) / norm1**2 
    
    rvec_2 = ri_atom - ri_2 # relative position between body and atom 2
    a_2 = 0
    if round(np.linalg.norm(rvec_2),10) != 0:
        norm2 = np.linalg.norm(rvec_2)
        # a due to F from atom 2
        a_2 = -( 2*(norm2)**(-12) - (norm2)**(-6) )*(rvec_2) / norm2**2
    
    a = a_1 + a_2
    
    return a

def LJ(r1, r2, v1, v2):

    for i in range(len(t)-1):
        # atom 1
        ai_1 = a(r1[i], r1[i], r2[i])
        r1[i+1] = r1[i] + dt*v1[i] + 0.5*(dt**2)*ai_1
        
        # atom 2
        ai_2 = a(r2[i], r1[i], r2[i])
        r2[i+1] = r2[i] + dt*v2[i] + 0.5*(dt**2)*ai_2
        
        # (atom 1) + 1
        aip1_1 = a(r1[i+1], r1[i+1], r2[i+1])
        v1[i+1] = v1[i] + 0.5*dt*(ai_1 + aip1_1)
        
        # (atom 2) + 1
        aip1_2 = a(r2[i+1], r1[i+1], r2[i+1])
        v2[i+1] = v2[i] + 0.5*dt*(ai_2 + aip1_2)
```

### Exercise 3, Lagrangian formalism and the Lennard-Jones potential (35pt)

The relevant chapters from Taylor are chapters 6-7 and 8. In addition he lectures
notes on the [Lagrangian Formalism and Calculus of
Variations](https://mhjensen.github.io/Physics321/doc/LectureNotes/_build/html/chapter8.html) can be of use. You
need also results from the two-body problem of chapter 8 of Taylor and
the lecture notes [two-body
problems](https://mhjensen.github.io/Physics321/doc/LectureNotes/_build/html/chapter6.html).

We will stay with a two-body problem only with particles $1$ and $2$, as in the two previous exercises.  
We introduce the relative mass $\mu=m_1m_2/M$ where the total mass $M=m_1+m_2$, the sum of the masses $m_1$ and $m_2$.

**3a (5pt):** Define the center-of-mass position $\boldsymbol{R}$ and the relative position $\boldsymbol{r}$ in terms of the masses and the positions of particles one and two $\boldsymbol{r}_1$ and $\boldsymbol{r}_2$, respectively. Define then the center-of-mass frame by setting $\boldsymbol{R}=0$. Show then that you can write the total angular momentum as $\boldsymbol{L}=\mu(\boldsymbol{r}\times \dot{\boldsymbol{r}})$. Hint: See Taylor section 8.3. How do you interpret the angular momentum?  Why can we reduce the motion of two particles to a problem in two dimensions only?

* The center-of-mass position $\boldsymbol{R}$ for particles 1 and 2 with total mass $M$ can be defined as:

$$
\boldsymbol{R} = 
\frac{m_1\boldsymbol{r}_1 + m_2\boldsymbol{r}_2}{M}
$$

* The relative position $\boldsymbol{r}$ for particles 1 and 2 is:

$$
\boldsymbol{r} = \boldsymbol{r}_1 - \boldsymbol{r}_2
$$

* When $\boldsymbol{R} = 0$,

$$
\frac{m_1\boldsymbol{r}_1 + m_2\boldsymbol{r}_2}{M} = 0
\;\;\; \rightarrow \;\;\;
\frac{m_1}{M}\boldsymbol{r}_1 = -\frac{m_2}{M}\boldsymbol{r}_2
$$

* Using the expression defined for relative position $\boldsymbol{r}$, the coordinates of the two particles can be expressed as:

$$
\boldsymbol{r}_1 = \frac{m_2}{M}\boldsymbol{r} \;\;\;\;\;\;\;\; \boldsymbol{r}_2 = -\frac{m_1}{M}\boldsymbol{r}
$$

* The angular momentum of any particle $i$ is $\boldsymbol{L} = \boldsymbol{r}_i \times \boldsymbol{p}_i$, so the sum the two particles' angular momenta is the total angular momentum of the system:

$$
\boldsymbol{L} 
\;\; = \;\; 
(\boldsymbol{r}_1 \times \boldsymbol{p}_1) + (\boldsymbol{r}_2 \times \boldsymbol{p}_2)
\;\; = \;\;
(\boldsymbol{r}_1 \times m_1\dot{\boldsymbol{r}}_1) + (\boldsymbol{r}_2 \times m_2\dot{\boldsymbol{r}}_2)
$$

* Substituting the derived forms of $\boldsymbol{r}_1$ and $\boldsymbol{r}_2$, total angular momentum can be written: 

$$
\boldsymbol{L} 
\;=\;
\frac{m_1 m_2}{M} \left( m_2\boldsymbol{r} \times \dot{\boldsymbol{r}} + m_1\boldsymbol{r} \times \dot{\boldsymbol{r}} \right)
$$

$$
\;=\;
\frac{m_1 m_2}{M^2} \left( m_2 + m_1 \right) (\boldsymbol{r} \times \dot{\boldsymbol{r}})
\;=\;
\mu(\boldsymbol{r}\times \dot{\boldsymbol{r}})
$$

* Angular momentum can be interpreted as the same as a single particle with mass $\mu$ and position (relative to the origin) $\boldsymbol{r}$. Because angular momentum is conserved, the cross product between $\boldsymbol{r}$ and the velocity $\dot{\boldsymbol{r}}$ must be constant, implying that the direction of the result is constant. These two vectors form a 2-dimensional plane, which justifies the reduction of the two-body problem to only 2 dimensions.

**3b (10pt):** We introduce polar coordinates with $r\in [0,\infty)$ and $\phi\in[0,2\pi]$. The quantity $r$ is the absolute value (magnitude) of the relative distance. Show that you can write the kinetic energy as $K=1/2\mu(\dot{r}^2+r^2\dot{\phi}^2)$.

* In Cartesian coordinates:
    * $ r = \sqrt{r_x^2 + r_y^2} $ where $r_x$ is the component of the relative position in the x-direction and $r_y$ is the component in the y-direction.
        * $ r_x = r\cos{(\phi)} $ and $ r_y = r\sin{(\phi)} $
    * $ \dot{r} =  \sqrt{\dot{r}_x^2 + \dot{r}_y^2}$

* Time derivatives of position using polar coordinates:

$$ 
\dot{r}_x = 
\frac{d}{dt}(\;r\cos{(\phi(t))}\;) = 
-r\sin{(\phi)}\dot{\phi} + \dot{r}\cos{(\phi)}
$$

$$ 
\dot{r}_y = 
\frac{d}{dt}(\;r\sin{(\phi(t))}\;) = 
-r\cos{(\phi)}\dot{\phi} + \dot{r}\sin{(\phi)}
$$

* Kinetic energy, Cartesian to polar:

$$
K = \frac{1}{2}\mu\dot{r}^2
$$

$$
\dot{r}^2 = \dot{r}_x^2 + \dot{r}_y^2 = (\; -r\sin{(\phi)}\dot{\phi} + \dot{r}\cos{(\phi)} \;)^2 + (\; -r\cos{(\phi)}\dot{\phi} + \dot{r}\sin{(\phi)} \;)^2
$$

$$
= r^2\dot{\phi}(\cos^2{(\phi)} + \sin^2{(\phi)}) + \dot{r}^2 (\cos^2{(\phi)} + \sin^2{(\phi)})
$$

$$
K = \frac{1}{2}\mu (\dot{r}^2 + r^2\dot{\phi})
$$

**3c (5pt):** Include the potential energy for the Lennard-Jones potential and write out the Lagrangian using the coordinates $r$ and $\phi$. Does the potential energy depend on $\phi$?

$$
\mathcal{L}(r, \dot{r}, \phi, \dot{\phi}, t) \;=\; 
K - V \;= \;
\frac{1}{2}\mu (\dot{r}^2 + r^2\dot{\phi}) - 4\varepsilon \left(\left(\frac{\sigma}{r}\right)^{12} - \left(\frac{\sigma}{r}\right)^6\right)
$$

* Potential energy does not depend on $\phi$.

**3d (10pt):** Use the Euler-Lagrange equations to find the equations of motion for $r$ and $\phi$. Does the equation of motion for $r$ agree with what you derived in exercise 1? Comment your results.

$$
\frac{\partial \mathcal{L}}{\partial r} - \frac{d}{dt}\frac{\partial \mathcal{L}}{\partial \dot{r}} = 0
$$

$$
= \mu r \dot{\phi}^2 + \frac{24\varepsilon}{r}\left(\left(\frac{\sigma}{r}\right)^{12} - \left(\frac{\sigma}{r}\right)^6\right) - \mu \ddot{r}

$$
\ddot{r} =  r \dot{\phi}^2 + \frac{24\varepsilon}{\mu r}\left(\left(\frac{\sigma}{r}\right)^{12} - \left(\frac{\sigma}{r}\right)^6\right)
$$

* Calculating the $\phi$-equation from the Lagrangian:

$$
