```python
import numpy as np
import pandas as pd
from math import *
import matplotlib.pyplot as plt
import os
```

Part 1

```python
d = 1.0 # m  
m = 1 #kg
delt = 0.01 # change in time   MHJ I changed also the time step, which was very large
tf = 30 #sec time final
n = ceil(tf/delt)
V_0 = 1 # J

# set up arrays for t, a, v, and y and we can compare our results with analytical ones
t = np.zeros(n)
v = np.zeros(n)
x = np.zeros(n)
 
# Initial conditions
def particles(v_0, x_0):
    x_0 = np.array([d])
    v_0 = np.array([v_0])
    x[0] = x_0  #m/s
    v[0] = v_0 #m

    for i in range(n-1):
    
        a = (-4 * V_0 / (m* d**4)) * (x[i]**3 - x[i]* d**2) # acceleration calculated from V(x)
   
        # update velocity and position
        v[i+1] = v[i] + delt * a
        x[i+1] = x[i] + delt * v[i+1]
        t[i+1] = t[i] + delt


    fig, ax = plt.subplots()
    ax.set_xlim(0 , tf)
    ax.set_xlabel('time')
    ax.set_ylabel('x meters')
    ax.plot(t, x)
    fig.tight_layout()
    plt.show()

p1 = particles(v_0 = 0.5, x_0 = d)
p2 = particles(v_0 = 1.5, x_0 = d)
p3 = particles(v_0 = 2.5, x_0 = d)
```
